```python
import numpy as np
from matplotlib import pyplot as plt
import matplotlib

# Create artificial data, 1000 samples
x = np.random.uniform(low=-3, high=3, size=1000)
y = []
for i in x:
    y.append(np.random.uniform(low=2*i-0.5, high=2*i+0.5, size=1)[0])

# alpha = 1.5
M = [i for i in range(1, 1001)]
alpha = 1.5
R = []

for i in M:
    r = 0
    for j in range(0, i):
        r += (y[j]-alpha*x[j])**2
    r/=i
    R.append(r)

# alpha = 2
alpha = 2
R2 = []
for i in M:
    r = 0
    for j in range(0, i):
        r += (y[j]-alpha*x[j])**2
    r/=i
    R2.append(r)

plt.plot(M, R, label='alpha = 1.5')
plt.plot(M, R2, label='alpha = 2')
plt.xlabel('M')
plt.ylabel('Risk')
plt.legend()

matplotlib.rcParams['figure.dpi'] = 300
```
